```python
import cv2
from PIL import Image
from matplotlib import pyplot as plt
```

```python
# Exemplo usando OpenCV:
import cv2
# Carregar uma imagem usando OpenCV
imagem_opencv = cv2.imread('img/imagem.jpg')

print(imagem_opencv)

# Exemplo usando PIL:
from PIL import Image
# Abrir uma imagem usando PIL
imagem_pil = Image.open('img/imagem.jpg')

print(imagem_pil)

#Mostrar a imagem
plt.imshow(imagem_opencv)
plt.axis('off')
plt.show
```

```python
# Redimensionar a imagem usando OpenCV
imagem_redimensionada_opencv = cv2.resize(imagem_opencv,(150,100))

#Mostrar a imagem
plt.imshow(imagem_redimensionada_opencv)
plt.axis('off')
plt.show
```

```python
#Mostrar a imagem
plt.imshow(imagem_pil)
plt.axis('off')
plt.show
```

```python
# Exemplo de converso de formato usando OpenCV:
# Converter a imagem para outro formato usando OpenCV
cv2.imwrite('img/imagem_conversao.png', imagem_opencv)

# Exemplo de converso de formato usando PIL:
# Converter a imagem para outro formato usando PIL
imagem_pil.save('img/imagem_conversao.png', format='PNG')
```

```python
imagem_opencv_2 = cv2.imread('img/imagem_cores_fracas.png')

#Mostrar a imagem
plt.imshow(imagem_opencv_2)
plt.axis('off')
plt.show
```

```python
# Carregar a imagem em escala de cinza
imagem = cv2.imread('img/imagem.jpg', cv2.IMREAD_GRAYSCALE)

# Equalizar o histograma da imagem
imagem_equalizada = cv2.equalizeHist(imagem)

plt.imshow(imagem)
plt.axis('off')
plt.show()

# Mostrar a imagem segmentada
plt.imshow(imagem_equalizada)
plt.axis('off')
plt.show()
```
